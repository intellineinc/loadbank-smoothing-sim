```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

index = count()


def animate(i):
    data = pd.read_csv('data.csv')
    even = data["cycle"] % 2 == 0
    x = data.loc[even, 'cycle']
    y1 = data.loc[even, 'tpo']
    y2 = data.loc[even, 'rpo']
    plt.cla()

    plt.plot(x, y1, label='Total Required Power Output')
    plt.plot(x, y2, label='Reported Power Output')

    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=500)

plt.tight_layout()
plt.show()
```
